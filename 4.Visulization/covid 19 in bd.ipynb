```python
import matplotlib.pyplot as pt
df = df.head(40)
pt.scatter(df["total_confirmed"],df["total_recovered"],color = "blue",label="scatter")
pt.xlabel("total_confirmed",color ="green")
pt.ylabel("total_recovered",color ="green")
pt.show()
```
