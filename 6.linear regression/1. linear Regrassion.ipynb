# 4 train ,test

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['area']]
y = df ['price']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.30)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

# 5 Create linear regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(xtrain,ytrain)
```

```python
reg.score(xtest,ytest)
```

```python
plt.xlabel('area in squarefeet')
plt.ylabel('Price in taka')
plt.scatter(df['area'],df['price'],marker='+',color = 'blue')
plt.plot(df.area,reg.predict(df[['area']]),color = 'red')
```

# 6. experiment by other value

```python
n = input('Enter the area:')
array = np.array(n)
array2 = array.astype(np.float)
value=([[array2]])
result = reg.predict(value)

home_price = np.array(result)
home_price = home_price.item()
print("Predicted our home price:",home_price)
```
